# **Dataset Analysis - Visualizing Earnings Based On College Majors**

I'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data, and cleaned by FiveThirtyEight (released on <a href="https://github.com/fivethirtyeight/data/tree/master/college-majors" target ="_blank">Github repo </a>). Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

* Rank - Rank by median earnings (the dataset is ordered by this column). 
* Major_code - Major code. 
* Major - Major description.
* Major_category - Category of major.
* Total - Total number of people with major.
* Sample_size - Sample size (unweighted) of full-time.
* Men - Male graduates.
* Women - Female graduates.
* ShareWomen - Women as share of total.
* Employed - Number employed.
* Median - Median salary of full-time, year-round workers.
* Low_wage_jobs - Number in low-wage service jobs.
* Full_time - Number employed 35 hours or more.
* Part_time - Number employed less than 35 hours.

Using visualizations, I will explore the following questions from the dataset :

## **Visualisation with Histograms**

We will explore the distributions of values in columns using histograms. 

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen"]
fig = plt.figure(figsize=(7,20))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist')
    ax.set_title(cols[r])
```

Based on the data visualization above, we can make a conclusion that the most common median salary is in a range between 30,000 USD - 40,000 USD.
Also, there are at least 28 majors (16.27% of total in the given dataset) that have predominatly female students (70% women share)and 24 majors (13.95% of total) with 60% of female students. 

```python
cols = ["ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(7,20))
for r in range(1,4):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist')
    ax.set_title(cols[r])
```

The most common unemployment rate based on the first histogram above is between 0.06 and 0.07. Moreover, there are overall more female graduates compared to male graduates of the given dataset.

## **Visualisation with Bar Plots**

In this section I will explore data using bar plots to compare the percentages of women (ShareWomen) from the first ten rows and last ten rows of the recent_grads dataframe, as well as  the unemployment rate (Unemployment_rate) from the first ten rows and last ten rows of the recent_grads dataframe.

```python
recent_grads.head(10).tail(10).plot.bar(x='Major', y='ShareWomen')
```

The chart above shows that some majors such as Astronomy and Astrophysics, and Actuarial Science have the highest women share among others, approximately 45% and 52% respectively. We can also observe that other majors are predominantly male, including all the engineering ones. However, we should remind that this chart only reprents the top and bottom 10 rows, we can't make a conclusion for the whole population. But knowing that the rows of the dataframe is ranked based on the medium salaries, the engineering majors are the first top majors in the dataset, thus we can conclude that men graduates are earning more than women on average.

```python
recent_grads.head(10).tail(10).plot.bar(x='Major', y='Unemployment_rate', legend =False)
```

The highest unemployment rate from our selected data is Nuclear Engineering major with only 0.18. The overall level of the unemployment rate is quite low, and, surprisingly, majors with the least medium salary are having lower unemployment compared to engineering majors.

## **Visualisation with Grouped Bar Plots**

In order to compare the number of men with the number of women in each category of majors we will use the grouped bar plot.

```python
recent_grads.groupby('Major_category')['Men', 'Women'].sum().plot(kind='bar')
```

The highest number of men and women are presented in Business majors, while the least amount is in Interdisciplinary majors. The predominantly female majors are Education, Health, abd Psychology & Social Work majors. Predominantly male majors are Computer & Mathematics and Engineering.
